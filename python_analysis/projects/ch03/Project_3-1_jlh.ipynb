```python
# 1 - Read the data from the CSV file into a DataFrame and display the first five rows.
import pandas as pd
ramen_data = pd.read_csv('ramen-ratings.csv')

# I like to include the basic information as well as a simple look at the data.
ramen_data.info()

# Here are the first five lines.
ramen_data.head()
```

```python
# 2 - Create a histogram for the Stars column using the number of bins necessary to 
#     display the ratings for each quarter of a Star.
ramen_data.plot.hist(y = 'Stars', bins = 20, title = 'Histogram of Stars data', legend = False);
```

```python
# 3 -Create a KDE plot for the Stars column.
ramen_data.plot.kde(y = 'Stars', title = 'KDE plot of Stars data', legend = False, grid = True, xlim = (-1, 6));
```

```python
# 4 -Create a bar plot that shows the average number of stars by ramen style.
#    Hint: You'll need to create the data before you plot it.

ramen_data.groupby('Style')['Stars'].mean().plot.bar(title = 'Bar plot by average number of stars.\nGrouped by Style', rot = 45, ylabel = 'Mean Value',\
                                                    ylim = (3, 5.5), grid = True);
```

```python
# 5 - Create a bar plot that shows the total number of reviews for each ramen type.

ramen_data.groupby('Style')['Stars'].count().plot.bar(rot = 45, ylabel = 'Count of reviews', 
                                                     title = 'Bar plot of total number of reviews.\nGrouped by Style');
```

```python
# 6 - Create a bar plot that shows the highest score give to each ramen type.

ramen_data.groupby('Style')['Stars'].max().plot.bar(rot = 45, ylabel = 'Max review valuess', 
                                                     title = 'Bar plot of highest score for a review.\nGrouped by Style');
```
