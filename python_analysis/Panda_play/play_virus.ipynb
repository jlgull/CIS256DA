# Pandas / matplotlib play

```python
import matplotlib.pyplot as mplt
%matplotlib inline
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data_2022-Sep-15.csv')
df.head(3)
```

```python
Sorted_df = df.sort_values(by = 'date', ascending = True)
Sorted_df.head(3)
```

```python
x = Sorted_df['date']
x.head(5)
```

```python
y = Sorted_df['newCasesBySpecimenDate'].round(5)
y.head(5)
```

```python
#%matplotlib inline
import mplcursors
mplt.plot(x,y)

mplcursors.cursor(hover = True)
mplt.show()
```

```python
%matplotlib inline
import mplcursors
fig = mplt.figure(figsize=(18,8))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_title('Covid cases in UK')
mplcursors.cursor(hover = True)
mplt.show()
```

```python
fig.savefig('plotcovid.png')
```
