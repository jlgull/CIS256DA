```python
# Task 01 Read the data from the CSV file into a DataFrame.
avocado_data = pd.read_csv("avocado.csv")
```

```python
# Task 05. Display the first and last five rows of data and the first and last four columns of data.
with pd.option_context(
        'display.max_rows', 10,
        'display.max_columns', 8):
    display(avocado_data)
```

```python
# Task 06. Choose any three columns, access them with bracket notation, and display the first five rows of this data.
avocado_data[['Date', 'Total Volume', 'region']].head(5)
```

```python
# Task 07. Select one column and access it with dot notation.
avocado_data.region
```

```python
# Task 08. Multiply the Total Volume and AveragePrice columns, and store the result in a new column called EstimatedRevenue.
#          Then, display the first five rows of this data to confirm that the column was added and has the correct values.
avocado_data['EstimatedRevenue'] = avocado_data['Total Volume'] * avocado_data.AveragePrice
#          Rather than display all 15 columns, I chose to only display the calculation and the 2 fields needed
#             for the calculation and the Date column.
avocado_data[['Date', 'Total Volume', 'AveragePrice', 'EstimatedRevenue']].head(5)
```

```python
# Task 09. Create a DataFrame thatâ€™s grouped by region and type and that includes the average price for the grouped columns.
#          Then, reset the index and display the first five rows.

avocado_data_new = avocado_data[['region','type','AveragePrice']].groupby(['region', 'type']).mean().copy()
# One of may trys to get the result, I think I have been asked for.
# avocado_data_new.groupby(['region', 'type']).mean().head()

avocado_data_new.reset_index(inplace = True)
avocado_data_new.head(5)

```

```python
# Task 10. Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year.
avocado_data.groupby('year')['Total Volume'].agg(['mean', 'median', 'std']).plot.barh()
```
