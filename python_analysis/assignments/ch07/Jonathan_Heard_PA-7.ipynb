## Import required modules

```python
# Import the modules needed for this project.

# Import pandas and assign the alias pd
import pandas as pd
```

# Tasks list

```python
# Task 01 Read the data from the CSV file into a DataFrame and display the first five rows.
avocado_data = pd.read_csv("avocado.csv")

# I wanted to have the basic information as well as the first five rows of data.
#    The order is critical, info() must come before head().

avocado_data.info(memory_usage = 'deep')

avocado_data.head(5)
```

```python
# Display the column names, before performing Task 2.

avocado_data.columns
```

```python
# Task 02 Notice that the column names are inconsistent. Change the column names so they all use title case with no spaces.

# Remove the spaces from all the column names with the replace() method and then apply the title() method.
avocado_data.columns = avocado_data.columns.str.replace(' ', '').str.title()
```

```python
# Validate that the column names have been cleaned up.

avocado_data.columns
```

```python
# Task 03 Add a column with the percentage of the bags in the TotalBags column that are extra-large bags.

avocado_data['pct_Totalbags_Xlarge'] = ((avocado_data.Xlargebags / avocado_data.Totalbags) * 100).round(2)
```

```python
# Task 04 Add a column with the percentage of the bags in the TotalBags column that are large bags.

avocado_data['pct_Totalbags_Large'] = ((avocado_data.Largebags / avocado_data.Totalbags) * 100).round(2)
```

```python
# Task 05 Add a column with the percentage of the bags in the TotalBags column that are small bags.

avocado_data['pct_Totalbags_Small'] = ((avocado_data.Smallbags / avocado_data.Totalbags) * 100).round(2)
```

```python
# Task 06 Display the first five rows of data to view the new columns.

avocado_data.head()
```

```python
# Task 07 Assign the Region, Type, Year, and TotalBags columns to a new DataFrame.

# Creat the new DataFrame.
avocado_data_new = avocado_data[['Region', 'Type', 'Year', 'Totalbags']].copy()

# Display the first 5 rows of the new DataFrame
avocado_data_new.head()
```

```python
# Task 08 Group, sum, and unstack this data to produce a new DataFrame that looks like this:

avocado_data_new = avocado_data_new.groupby(['Region', 'Year', 'Type']).sum().unstack(level = 'Type')

# display the final results
avocado_data_new
```
