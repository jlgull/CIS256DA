```python
## Import all the required modules, pandas and seaborn.
```

```python
import pandas as pd
import seaborn as sns
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

## Questions 9 through 14

```python
# Question 9 - Vertical bar plot

sns.catplot(data = mortality_data.query('Year in (1900, 1950, 2000)'),
            kind = 'bar', x = 'AgeGroup', y = 'DeathRate');
```

```python
# Question 9 - Vertical bar plot, show death rate per year and set aspect ratio to 1.8 - part 2

sns.catplot(data = mortality_data.query('Year in (1900, 1950, 2000)'),
            kind = 'bar', x = 'AgeGroup', y = 'DeathRate', hue = 'Year', aspect = 1.8);
```

```python
# Question 10 - same as Question 9; but using subplots

sns.catplot(data = mortality_data.query('Year in (1900, 1950, 2000)'),
            kind = 'bar', x = 'AgeGroup', y = 'DeathRate', height = 4,
           col = 'Year', col_wrap = 3);
```

```python
# Question 11 - using a specific method, draw a line plot for the 15-19 age group

sns.lineplot(data = mortality_data.query('AgeGroup == "05-09 Years"'), x = 'Year', y = 'DeathRate');
```

```python
# Question 11 - using a specific method, draw a line plot for the 15-19 age group, add title
#   and modify y-axis label using Axes methods - part 2

ax = sns.lineplot(data = mortality_data.query('AgeGroup == "05-09 Years"'), x = 'Year', y = 'DeathRate')

ax.set_title('Death rate for the 05-09 Age Group')

ax.set_ylabel('Deaths per 100,000');
```

```python
# Question 11 - using a specific method, draw a line plot for the 15-19 age group, add title
#   and modify y-axis label using Axes methods - part 2, alternate approach.

ax = sns.lineplot(data = mortality_data.query('AgeGroup == "05-09 Years"'), x = 'Year', y = 'DeathRate')

ax.set(title = 'Death rate for the 05-09 Age Group', ylabel = 'Deaths per 100,000');
```

```python
# Question 12 - line plot for years 1950 to 2000

sns.relplot(data = mortality_data.query('Year >= 1950 and Year <= 2000'), kind = 'line',
            x = 'Year', y = 'DeathRate', hue = 'AgeGroup');
```

```python
# Quetion 13 - Scatter plot of data in Question 12

sns.relplot(data = mortality_data.query('Year >= 1950 and Year <= 2000'), kind = 'scatter',
            x = 'Year', y = 'DeathRate', hue = 'AgeGroup',
           size = 'DeathRate', sizes = (10, 100));

```

```python
# Question 14 - create 4 bar subplots, 2 x 2

g = sns.catplot(data = mortality_data.query('Year in (1900, 1925, 1950, 1975, 2000)'),
                kind = 'bar', x = 'Year', y = 'DeathRate', hue = 'AgeGroup', legend = False,
                col = 'AgeGroup', col_wrap = 2, height = 4)
g.fig.suptitle('Subplot of Death Rate by Age')
for ax in g.axes.flat:
    ax.set_ylabel('Deaths per 100,000');
```

```python
# Question 14 - create 4 bar subplots, 2 x 2, save as png.

g = sns.catplot(data = mortality_data.query('Year in (1900, 1925, 1950, 1975, 2000)'),
                kind = 'bar', x = 'Year', y = 'DeathRate', hue = 'AgeGroup', legend = False,
                col = 'AgeGroup', col_wrap = 2, height = 4)
g.fig.suptitle('Subplot of Death Rate by Age', y = 1.025)
for ax in g.axes.flat:
    ax.set_ylabel('Deaths per 100,000');
g.savefig('barCharts.png')
```
