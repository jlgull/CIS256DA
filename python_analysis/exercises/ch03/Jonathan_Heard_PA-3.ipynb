```python
# The only imported module needed for this project is pandas.
import pandas as pd
```

# Tasks list

```python
# Task 01 Read the data from the .csv file into a DataFrame and display the first five rows.
avocado_data_raw = pd.read_csv("avocado.csv")

# I decided I wanted to have the basic information as well as the first five rows of data.
#    What if found was, I had to do it this order. If I put '.head()' first, it did not diplay.
#    Could have used two cells; but I was trying to keep it to one.
avocado_data_raw.info()
avocado_data_raw.head(5)
```

```python
# Task 02 Create a new DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags columns grouped by type, and then display the DataFrame.

avocado_data_new = avocado_data_raw.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum().head()

# Having a desire to formte the totals data, did a Duck-Duck-go Search.
# Found this as a working answer on this site: 
#                 https://stackoverflow.com/questions/20937538/how-to-display-pandas-dataframe-of-floats-using-a-format-string-for-columns
pd.options.display.float_format = '{:,.2f}'.format

# The order for doing both .info() and .head() matters. Reversing the order give no .head() results.
print('\nDisplaying the ".info()" and ".head()"for the new DataFrame.\n')
avocado_data_new.info()

avocado_data_new.head()

```

```python
# Task 03 Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado.
avocado_data_new.plot.bar(title = 'Bar plot of the various sizes of bags', ylabel = 'Count of Bags', rot = 45 );
```

```python
# Task 04. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.

avocado_data_raw.plot.scatter(x = 'Total Volume', y = 'AveragePrice');

# Not sure which order to do the scatter plot, so I did both. I can't say either one actully tells me
#     anything; but here is the order from the question.
```

```python
# Task 04a. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.

# Here is the order switched, still not sure which one actually tells the viewer something useful.

avocado_data_raw.plot.scatter(x = 'AveragePrice', y = 'Total Volume');
```

# Questions

```python
# 1. In what year were the most bags of avocados sold?

avocado_data_raw.groupby('year')['Total Bags'].agg(sum).plot.line(x = 'year', grid = True, title = '\nThe most bags of Avocados were sold in 2017.\n',\
                                                                 ylabel = 'Total bags per year.', rot = 45
                                                                 );
```

```python
# 2. Do the total bags of avocados appear to have an effect on the average price?

#    After playing with the data and trying various plot methods, the scatter plot
#      seemed the show the data best. I did use ylim, during testing; but with no real
#      obvious benefit. I even tried putting 2 kde plots on the same graph; all it showed 
#      was 2 lines that didn't intersect.

avocado_data_raw.plot.scatter(x = 'AveragePrice', y = 'Total Bags', \
                              title = 'Based on the data presented,\nthe lower the total bag count,\nthe higher the average price.', \
                             figsize = (8, 8));
```
