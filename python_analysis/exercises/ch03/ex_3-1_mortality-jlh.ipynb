# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate');
```

```python
mortality_wide.plot();
```

```python
# 3 - Use wide data to create a line plot for 15-19 Years data, with title and no legend.
mortality_wide.plot.line(y = '15-19 Years', legend = False, title = 'Line Plot of "15-19 Years" data.');
```

```python
# 4 - Use wide data to create an area plot for all age groups and reverse the legend order
mortality_wide.plot.area(title = 'Area plot for all age groups', legend = 'reverse');
```

```python
# 5 - Use wide data to create an bar plot for all age groups mortality for just 1900.
mortality_wide.query('Year == 1900').plot.bar(title = 'Bar plot for the Year 1900 only.', rot = 45);
```

```python
# 6 - Use wide data to create an bar plot for all age groups mortality for just 1900.
mortality_wide.query('Year == 2000').plot.bar(title = 'Bar plot for the Year 2000 only.', figsize = (8, 4), \
                                              rot = 45, xlabel = '', ylim = (0, 100), ylabel = 'Mortality rate per 100,000');
```

```python
# 7 - Use long data to create a bar similar to step # 6.
mortality_data.query('Year == 2000').pivot(index = 'AgeGroup', columns = 'Year', values = 'DeathRate')\
              .plot.bar(title = 'Bar chart for the Year 2000 only', figsize = (8, 4), legend = False,\
                        rot = 45, ylabel = 'Mortality rate per 100,000', xlabel = '');
```

```python
# 8 - Use long data to create histogram showing the history od death rates in the defaults number of bins (10)
mortality_data.plot.hist(y = 'DeathRate', title = 'Histogram of Death Rate',\
                        rot = 45, legend = False );
```

```python
# 8 - Use long data to create histogram showing the history of death rates with bins set to 15.
mortality_data.plot.hist(y = 'DeathRate', title = 'Histogram of Death Rate', bins = 15,\
                        rot = 45, legend = False );
```

```python
# 9 - Use long data to create a density plot, for the year 1900
mortality_data.query('Year == 1900').plot.density(y = 'DeathRate', title = 'Density ploy of Death Rate for 1900', legend = False, grid = True);
```

```python
# 10 - Use long data to create a pie chart to show sum of death rate
mortality_data.query('Year in [1900, 1925, 1950, 1975, 2000]').groupby('Year')['DeathRate'].sum()\
              .plot.pie(title = 'Pie Chart for Sum of Death Rate\n for the every 25 years from 1900 to 2000.', ylabel = '', figsize = (8,8),\
                       autopct = '%1.2f%%') ;
```

```python
print(f'\n\tThis display is a subplot for all the data, years 1900 to 2000\n')
mortality_wide.plot.line(
    title = ['Child Mortality: 01-04', 'Child Mortality: 05-09', 'Child Mortality: 10-14', 'Child Mortality: 15-19'],
    ylabel = 'Deaths per 100,000', sharey = True, subplots = True, layout = (2, 2), figsize = (10,10), legend = False, grid = True, rot = 45);
```

```python
# 11 - Create a plot with 4 subplots,of the Death Rate
print(f'\n\tThis display is a subplot for the years listed.\n')
mortality_wide.query('Year in [1900, 1925, 1950, 1975, 2000]').plot.barh(
    title = ['Child Mortality: 01-04', 'Child Mortality: 05-09', 
             'Child Mortality: 10-14', 'Child Mortality: 15-19'],\
    subplots = True, layout = (2, 2), ylabel = 'Year', sharey = True, \
    grid = True, legend = False, figsize = (8, 5), rot = 45);
```
