# Exercise 7-2: Prepare the Cars data

## Read the data

```python
import pandas as pd
```

```python
cars = pd.read_csv('cars.csv')
```

```python
# 4 
# display the first five rows

cars.head()
```

## Add and drop columns

```python
# 5 
# display the unique values for the CarName column

cars.CarName.unique()
```

```python
# 6 
# add a column for the brand by using a lambda expression, axis = 1 for rows.

cars['brand'] = cars.apply(lambda x: x.CarName.split(' ')[0], axis = 1)

cars.brand
```

```python
# display all the unique data from the new brand column

cars.brand.unique()
```

```python
# Replace all misspelled or inconsistant brand names

cars.brand.replace({'Nissan':'nissan', 'maxda':'mazda', 'porcshce':'porsche',
                   'vokswagen':'volkswagen', 'vw':'volkswagen'}, inplace = True)
```

```python
# display all the unique data from the cleaned up brand column

cars.brand.unique()
```

```python
# 7 
# add a column for the model by using a user-defined function, axis = 1 for rows.

def define_model(row):
    names = row.CarName.split(' ') # this builds the CarName split into 2 parts
    names = ' '.join(names[1:])    # this adds a space before the model number information.
    return names

cars['model'] = cars.apply(define_model, axis = 1)

cars.model
```

```python
# 8
# drop the CarName and car_ID columns

cars = cars.drop(columns = ['CarName', 'car_ID'])

cars.head(20)
```

## Set an index and unstack the data

```python
# 9
# filter the cars data and store in a variable named cars_filtered
# only get rows for Volkswagon and
# the columns named model, horsepower, carbody, and doornumber

cars_filtered = cars.query('brand == "volkswagen"')[['model', 'horsepower', 'carbody', 'doornumber']]

cars_filtered
```

```python
# 10
# index the cars_filtered data by model, carbody, and doornumber

cars_indexed = cars_filtered.set_index(['model', 'carbody', 'doornumber'])

cars_indexed
```

```python
# 11
# unstack the carbody column

cars_indexed.unstack('carbody')
```
