# Project 4-1 Plot ramen data with Seaborn

```python
# Import required modules

import pandas as pd
import seaborn as sbn
```

## Tasks

```python
# Task 1, Read the CSV data and display the first 5 rows

ramen_base_data = pd.read_csv('ramen-ratings.csv')
ramen_base_data.head(5)
```

```python
# Task 2, Create bar plot to show the Stars by Style, using a Seaborn specific method.

sbn.barplot(data = ramen_base_data, x = 'Style', y = 'Stars');
```

```python
# Task 3, Create bar plot to show the Stars by Style, using a Seaborn generic method.

sbn.catplot(data = ramen_base_data, kind = 'bar', x = 'Style', y = 'Stars');
```

```python
# Task 4, Modify Task 3 plot to use custom titles x-axis label

g = sbn.catplot(data = ramen_base_data, kind = 'bar', x = 'Style', y = 'Stars');

for ax in g.axes.flat:
    ax.set(title = 'Rating by Ramen Style.', xlabel = 'Ramen Style');
```

```python
# Task 5, Use the query() method to restrict country to Japan, India, Taiwan and U.S.
#         Plot the Stars by Style for each country on a subplot with 2 subplots per row.

g = sbn.catplot(data = ramen_base_data.query('Country in ["Japan", "India", "Taiwan", "USA"]'), 
                                       kind = 'bar', x = 'Style', y = 'Stars',
                                        col = 'Country', col_wrap = 2);
```

```python
# Task 6, Modify the Task 5 plot, to create custom title for each subplot.
#         Hint: Use the enumerate() method.

g = sbn.catplot(data = ramen_base_data.query('Country in ["Japan", "India", "Taiwan", "USA"]'), 
                                       kind = 'bar', x = 'Style', y = 'Stars',
                                        col = 'Country', col_wrap = 2);

countries = ramen_base_data.query('Country in ["Japan", "India", "Taiwan", "USA"]').Country.unique()

for i, ax in enumerate(g.axes.flat):
    ax.set(title = "Rating by Ramen Type in " + countries[i], xlabel = 'Ramen Type')

```

```python
# Task 6, Modify the Task 5 plot, to create custom title for each subplot.
#         Hint: Use the enumerate() method. Tried, with countries pre-defined.

countries = ("Japan", "India", "Taiwan", "USA")

g = sbn.catplot(data = ramen_base_data.query('Country in @countries'), 
                                       kind = 'bar', x = 'Style', y = 'Stars',
                                        col = 'Country', col_wrap = 2);

for i, ax in enumerate(g.axes.flat):
    ax.set(title = "Rating by Ramen Type in " + countries[i], xlabel = 'Ramen Type')

```
